**add two numbers**

```python
t=int(input())
while(t>0):
  a,b=input().split()
  a,b=int(a),int(b)
  print(a+b)
  t-=1
```

**find reminder**

```python
t=int(input())
while(t>0):
  a,b=input().split()
  a,b=int(a),int(b)
  print(a%b)
  t-=1
```

**sum of 1 and last digit**

```python
t=int(input())
while(t>0):
  a=input()
  f=int(a[0])
  l=int(a[-1])
  print(f+l)
  t-=1
```
